# 10 Constrastes de Independencia

Dado $n$ individuos se clasifican según dos criterios $(X,Y)$. Sean $x_1,\ldots, x_I$ los distintos niveles del criterio $X$ y sean $y_1,\ldots, y_J$ los distintos niveles del criterio $Y$. 

$\textbf{Ejemplo}$ Supongamos que tenemos una tabla de contingencia que muestra la relación entre el nivel educativo y el sector de empleo de una muestra de personas. Queremos determinar si el nivel educativo y el sector de empleo están independientemente relacionados.

```python
import numpy as np
import pandas as pd

datos = {'Nivel educativo': np.repeat(np.array(['Secundaria', 'Preparatoria', 'Universidad']), (60, 75, 55)),
         'Sector de empleo': np.repeat(np.array(['Independiente','Privado', 'Público', 
                                                 'Independiente','Privado', 'Público', 
                                                 'Independiente','Privado', 'Público']), (15, 25, 20, 25, 20, 30, 10, 15, 30))
        }
df = pd.DataFrame(datos)

tabla_contingencia1 = pd.crosstab(df['Nivel educativo'], df['Sector de empleo'], margins=True)

tabla_contingencia1
```

$\textbf{Ejemplo}$ En un estudio de una vacuna de hepatitis participan 1083 voluntarios. De éstos, se eligen aleatoriamente 549 y son vacunados. Los otros, 534, no son vacunados. Después de un cierto tiempo, se observa que 70 de los 534 no vacunados han contraído la hepatitis, mientras que sólo 11 de los 549 vacunados la han contraído.

```python
datos = {'Sintomas': np.repeat(np.array(['Si', 'No']), (81, 1002)),
         'Vacuna': np.repeat(np.array(['Con','Sin', 'Con','Sin']), (11, 70, 538, 464))
        }
df = pd.DataFrame(datos)

tabla_contingencia2 = pd.crosstab(df['Sintomas'], df['Vacuna'], margins=True)
tabla_contingencia2
```

Así, definimos $n_{ij}$ como el número de individuos clasificados en el nivel $x_i$ según el criterio $X$ y clasificados en el nivel $y_j$ según el criterio $Y$. A partir de dichos valores construimos la denominada tabla de contingencia:

<img src=tablacon.png>

En la tabla anterior, $n_{i∙}$ sería el número total de individuos clasificados en el nivel $x_i$ según el criterio $X$ y  $n_{∙j}$ el número total de individuos clasificados en el nivel $y_j$ según el criterio  $Y$.
 

De este modo las hipótesis serían las siguientes:

$H_0:$ Los criterios de clasificación $X$ y $Y$ son independientes,

$H_1:$ Los criterios de clasificación $X$ y $Y$ son dependientes,


Modelando el problema anterior como un test de bondad de ajuste se tendría que considerar una nueva variable $Z=(X,Y)$, denominada variable modelo, cuyo dominio es $\{(x_1,y_1), ..., (x_I, y_J)\}$.

Bajo la hipótesis $H_0$ se tiene que la distribución de $Z$ es:

<img src=probconj.png>

El estadístico asociado es:

<img src=chi-cuad.png>

el cual sigue aproximadamente una distribución $\chi^2$ con (I-1)(J-1) grados de libertad.

En la biblioteca scipy.stats contiene la función chi2_contingency para realizar estas operaciones.

```python
from scipy.stats import chi2_contingency

chi2_contingency(tabla_contingencia1)
```

Conclusión: No se rechaza la hipótesis nula.
No hay suficiente evidencia para afirmar que el nivel educativo y el sector de empleo son independientes.

```python
chi2_contingency(tabla_contingencia2)
```

Conclusión: Se rechaza la hipótesis nula.
