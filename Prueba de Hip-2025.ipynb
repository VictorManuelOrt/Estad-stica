# Prueba de Hipótesis

Una prueba de hipótesis es descrita por las siguientes componentes:
\begin{align}
(X_1, \ldots, X_n; H_0, H_1; G)
\end{align}
donde $X_1, \ldots, X_n$ es una muestra aleatoria, $H_0$ y $H_1$ son hipótesis concernientes a la distribución de probabilidad de la población, y $G$ es un subconjunto de $\mathbb{R}^n$.

El conjunto $G$ se denomina región crítica o de rechazo. Si $(X_1=x_1, X_2=x_2, \ldots, X_n=x_n)\in G\subset \mathbb{R}^n$ entonces se acepta $H_1$ en caso contrario se acepta $H_0$.

$\textbf{Ejemplo}$ Considere el siguiente juego de hipótesis
\begin{align}
H_0: X\sim N(500, 50)
\end{align}

\begin{align}
H_1: X\sim N(490, 50)
\end{align}

equivalentemente, $H_0:\mu=500$ vs $H_1:\mu=490$.

Regla de Decisión: Si $X_1, X_2\leq 496$ entonces aceptamos $H_1$, i.e. $G=\{(x_1, x_2)\in \mathbb{R}^2:x_1,x_2\leq 496\}$.

## Contrastes de hipótesis para el parámetro $\mu$ de una variable normal con $\sigma$ conocida

$H_0:\mu=\mu_0$ vs $H_1:\mu> \mu_0$

Estadístico:
\begin{align}
Z=\frac{\bar{X}-\mu}{\sigma/\sqrt{n}}\sim N(0,1)
\end{align}

La regla consistirá en rechazar $H_0$ si el estadístico de contraste $Z$ es mayor a un umbral ($z$) por determinar, el cual se obtiene vía el nivel de significancia ($\alpha$). 

\begin{align}
\alpha=P(Z>z)\Longleftrightarrow 1-\alpha=P(Z\leq z).
\end{align}

es decir, $z=z_{1-\alpha}$.

<img src="r1.png">

En conclusión, Rechazar $H_0$, si el estadístico de contraste
\begin{align}
Z_0=\frac{\bar{X}-\mu_0}{\sigma/\sqrt{n}}>z_{1-\alpha}
\end{align}


Observe que también queda determinado un intervalo de confianza para $\mu$:

$$Z<z_{1-\alpha}\Longrightarrow \frac{\bar{X}-\mu}{\sigma/\sqrt{n}}<z_{1-\alpha}$$

$$\mu\in (\bar{X}-z_{1-\alpha}\frac{\sigma}{\sqrt{n}}, \infty)$$

## El $p$-valor

El $p$-valor o valor crítico de un contraste es la probabilidad que el estadístico de contraste tome un un valor tan extremo o más que el que se ha observado, dado que $H_0$ es verdadera. En el caso anterior se tendría:

\begin{align}
p=\mathbb{P}(Z\geq Z_0|H_0)
\end{align}

##  El método de los seis pasos 

1. Establecer la hipótesis nula $H_0$ y la hipótesis alternativa $H_1$.

2. Fijar un nivel de significacia $\alpha$.

3. Seleccionar el estadístico de contraste apropiado.

4. Calcular el valor del estadístico de contraste a partir de los datos muestrales.

5. Calcular el $p$-valor del contraste.

6. Decisión: rechazar $H_0$ en favor de  $H_1$ si el $p$-valor es más pequeño que $\alpha$; en caso contrario, aceptar $H_0$.

## 8.2 Contrastes de hipótesis para el parámetro $\mu$ de una variable normal con $\sigma$ desconocida

En este caso debemos considerar al estadístico $T$:
$$ T=\frac{\bar{X}-\mu}{\frac{S}{\sqrt{n}}}\sim T-Student(n-1)$$

$\textbf{Ejemplo 8.2.1}$ Se espera que el nivel de colesterol en un grupo de pacientes bajo un determinado tratamiento se distribuya normalmente con media 220 mg/dl.

Se toma una muestra de 9 pacientes, y se miden sus niveles: 203, 229, 215, 220, 223, 233, 208, 228,209.
Contrastar la hipótesis que esta muestra efectivamente proviene de una población con media 220 mg/dl.
$H_0: \mu=220$ vs $H_1:\mu \neq 220$. 

```python
import numpy as np

colesterol = np.array([203, 229, 215, 220, 223, 233, 208, 228, 209])

media_muestral = np.mean(colesterol)
desv_muestral = np.std(colesterol, ddof=1)  
n = len(colesterol)
estadistico_contraste = (media_muestral - 220) / (desv_muestral / np.sqrt(n))

estadistico_contraste
```

```python
import scipy.stats as stats
2*(1-stats.t.cdf(abs(estadistico_contraste), df=8))
```

```python
from scipy.stats import ttest_1samp
result = ttest_1samp(colesterol, popmean=220, alternative='two-sided') 
result
```

```python
help(ttest_1samp)
```

$\textbf{Conclusión:}$ Como que el $p$-valor$>0.1$, no podemos rechazar que el nivel medio de colesterol en plasma sea igual a 220 mg/dl.

## Contrastes de hipótesis para el parámetro $\sigma$ de una variable con distribución normal

En el caso de hipótesis referente a la varianza (desviación estandar) de una población se utiliza el estadístico:
$$ \chi^2_{n-1}=\frac{(n-1)S^2}{\sigma^2}$$
el cual tiene una distribución $\chi^2$ con n-1 grados de libertad.

$\textbf{Ejemplo}$ Verifiquemos que si la varianza de la amplitud del sépalo de las flores de la base de datos iris es menor que 0.2. Considere un nivel de significancia de 0.05.
$H_0:\sigma^2= 0.2$ vs $H_1=\sigma^2< 0.2$

```python
from scipy.stats import chi2

flores_elegidas_chi = iris.sample(40, replace=False)

varianza_muestral = flores_elegidas_chi['sepal_width'].var()
n = len(flores_elegidas_chi)
varianza_hipotetica = 0.2
estadistico_prueba = (n - 1) * varianza_muestral / varianza_hipotetica

valor_critico = chi2.ppf(0.05, df=n-1)  
print(valor_critico)
print(estadistico_prueba)
```

Concluimos por tanto, que en base a la muestra no tenemos evidencias suficientes para aceptar que la varianza de la amplitud del sépalo sea menor que 0.2.

 ## Contrastes de hipótesis para dos muestras

En este contexto se busca comparar el valor de un parámetro en dos poblaciones. De este modo se requiere una muestra para cada población. En este caso las hipótesis alternativas son:

$\bullet$  las medias muestrales son distintas.

$\bullet$ la media de la primera muestra es mayor que la segunda.

$\bullet$ la media de la primera muestra es menor que la segunda.

```python
setosa=iris.query('species == "setosa"')
versicolor=iris.query('species == "versicolor"')
```

```python
muestra_setosa = setosa.sample(30, replace=True)
muestra_versicolor = versicolor.sample(30, replace=True)
```

```python
from scipy.stats import ttest_ind
t_statistic, p_value = ttest_ind(muestra_setosa['petal_length'], muestra_versicolor['petal_length'], alternative='two-sided')
print(p_value)
```
