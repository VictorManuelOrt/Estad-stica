### 4.1 Muestreo Estadístico

### $\bf{Población:}$ conjunto de sujetos/objetos con una o varias características que podemos medir. $\newline$

### $\bf{Muestra:}$ subconjunto de una población.

### Cuando la selección de una muestra de la población se realiza de manera que todas las muestras del mismo tamaño sean equiprobables, se denomina $\bf{muestreo}$ $\bf{aleaorio}$. 

### Consideremos una urna de 100 bolas numeradas del 1 al 100:

Para generar la muestra aleatoria en python usaremos la función $\bf{random.choices}$ del módulo $\bf{random}$.

```python
import random

esferas = list(range(1, 101))
muestra = random.choices(esferas, k=16)
muestra
```

Para el siguiente ejemplo usaremos una base de datos denominada iris, la cual importamos usando $\textbf{Seaborn}$. La tabla de datos iris contiene 150 flores de 3 especies diferentes: setosa, versicolor y virginica.

```python
import seaborn as sns
iris = sns.load_dataset('iris')
```

```python
iris
```

### Si quisiéramos una muestra de 20 flores con reposición, haríamos lo siguiente:

```python
muestra1 = iris.sample(20, replace=True)
muestra1
```

### Otra manera de extraer nuestra muestra sería repetir 15 veces el proceso de sacar una bola de la urna pero ahora sin devolverla. En este caso se habla de una $\bf{muestra}$ $\bf{aleatoria}$ $\bf{sin}$ $\bf{reposición}$.

### Para obtener una muestra de 10 flores sin reposición, haríamos los pasos siguientes:

```python
muestra2= iris.sample(20, replace=False)
muestra2
```

### Supongamos que los individuos de una población vienen dados en forma de una lista ordenada. El $\bf{muestreo}$ $\bf{sistemático}$ consiste en tomarlos a intervalos constantes escogiendo al azar el primer individuo que elegimos.

<img src="urna3.png">

### Ahora calculemos una muestra aleatoria sistemática de la tabla de datos iris de tamaño 10.

```python
import numpy as np 

primera_flor = random.randint(0, 151)
incremento = 15

flores_elegidas = np.arange(primera_flor, primera_flor + incremento * 10, incremento)
print(flores_elegidas)

flores_elegidas_sis=flores_elegidas%150
print(flores_elegidas_sis)
```

```python
muestra3=iris.iloc[flores_elegidas_sis, :]
muestra3
```

### Cuando la población está clasificada en estratos (grupos), se toma una muestra aleatoria de cada estrato y se unen en una muestra global. A este proceso se le llama $\bf{muestreo}$ $\bf{aleatorio}$ $\bf{estratificado}$.

<img src="urna4.png">
